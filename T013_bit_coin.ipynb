# Data Cleaning or Imputing
In statistics, imputation is the process of replacing missing data with substituted values. When substituting for a data point, it is known as "unit imputation". when substituting for a component of a data point, it is known as "item imputation". 
# So why imputation
    There are three main problems that missing data causes:
    missing data can introduce a substantial amount of bias, make the handling and analysis of the data more arduous, and create reductions in efficiency. Because missing data can create problems for analyzing data, imputation is seen as a way to avoid pitfalls involved with listwise deletion of cases that have missing values.

# Regression using StatsModels

Statsmodels is a library used widely for linear regression. The linear regression library has convenient ways to specify parameters for regression and hence is popularly used. The format to specify regression for a single variable is:

lm = sm.ols(formula = 'y ~ x')
where x is a dependent variable. Let us look at how market price is affected by mined bitcoins

lm = sm.ols(formula = 'btc_market_price ~ btc_total_bitcoins').fit()

This is called simple regression as only one variable affects the target variable where input is btc_total_bitcoins and target is btc_market_price 

Now get prediction of the model
    y_hat = lm.predict(my_dataframe[['btc_total_bitcoins']])

## Prediction and Improving the fit


If many variables are involved in modeling target, it is called multiple linear regression. The most important parameter to look for is mean_squared_error which tells us how good the predictions matched.

Fit can be improved using rsquared and p-values.

R-squared is a value between 0 and 1.0. The higher the value, the better is the fit.

P-values show significance of a variable in the model. P-values above 0.05 are considered insignificant.

Build a model of Market Price for input as all the features
Run the prediction on training model. Assign predictions to y_hat
What is MSE? What is r_squared?


```python
lm = sm.ols(formula='btc_market_price ~ Date + btc_n_orphaned_blocks + btc_total_bitcoins + btc_market_cap + btc_trade_volume + btc_blocks_size + btc_avg_block_size + btc_n_transactions_per_block + btc_median_confirmation_time + btc_hash_rate + btc_difficulty + btc_miners_revenue + btc_transaction_fees + btc_cost_per_transaction + btc_n_unique_addresses + btc_n_transactions + btc_n_transactions_total + btc_n_transactions_excluding_popular + btc_n_transactions_excluding_chains_longer_than_100 +btc_estimated_transaction_volume_usd + btc_output_volume + btc_estimated_transaction_volume ', data=my_dataframe).fit()
y_hat = lm.predict(my_dataframe[['Date','btc_n_orphaned_blocks','btc_total_bitcoins', 'btc_market_cap','btc_trade_volume', 'btc_blocks_size', 'btc_avg_block_size','btc_n_transactions_per_block', 'btc_median_confirmation_time','btc_hash_rate', 'btc_difficulty', 'btc_miners_revenue', 'btc_transaction_fees','btc_cost_per_transaction', 'btc_n_unique_addresses', 'btc_n_transactions', 'btc_n_transactions_total', 'btc_n_transactions_excluding_popular', 'btc_n_transactions_excluding_chains_longer_than_100','btc_estimated_transaction_volume_usd','btc_output_volume', 'btc_estimated_transaction_volume']])
print("MSE:", mean_squared_error(my_dataframe['btc_market_price'].values, y_hat))
print(lm.rsquared)
lm.summary()
```
